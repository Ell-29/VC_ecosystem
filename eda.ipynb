# EDA 
## Loading the data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 100)
```

```python
DATA_DIR = Path("data")
csv_files = list(DATA_DIR.glob("*.csv"))
csv_files
```

```python
dfs = {}

for file in csv_files:
    name = file.stem
    dfs[name] = pd.read_csv(file, low_memory=False)
    print(name, dfs[name].shape)
    
```

```python
for name, df in dfs.items():
    print("\n" + "-"*50)
    print(name)
    display(df.head())
```

## Missing values 

```python
for name, df in dfs.items():
    print("\n" + "="*50)
    print(name)

    missing = (
        df.isna()
        .mean()
        .mul(100)
        .sort_values(ascending=False)
    )

    missing = missing[missing > 0]
    display(missing)
```

### Exploring the main table: objects

```python
objects = dfs.get("objects")

objects.shape
```

```python
objects.head()
```

```python
objects.columns
```

```python
## creating a new table for the companies 

companies = objects.copy()

if "entity_type" in companies.columns:
    companies = companies[companies["entity_type"] == "Company"].copy()

companies.shape
```

```python
companies.head()
```

```python
date_cols = ["founded_at", "closed_at", "first_funding_at", "last_funding_at"]

for col in date_cols:
    if col in companies.columns:
        companies[col] = pd.to_datetime(companies[col], errors="coerce")

num_cols = ["funding_total_usd", "funding_rounds"]

for col in num_cols:
    if col in companies.columns:
        companies[col] = pd.to_numeric(companies[col], errors="coerce")

if "founded_at" in companies.columns:
    companies["founded_year"] = companies["founded_at"].dt.year
```

# Funded EDA

```python
funded = companies[
    (companies["funding_total_usd"] > 0) &
    (np.isfinite(companies["funding_total_usd"]))
].copy()
```

**Distribution of Total Funding**<br>
This graph shows how total funding is distributed across all funded startups. Since a small number of companies raise very large amounts, we use a log scale on the x-axis to make the distribution easier to read.

```python
plt.figure(figsize=(8,5))
sns.histplot(funded["funding_total_usd"], bins=50)
plt.xscale("log")
plt.title("Distribution of Total Funding")
plt.xlabel("Total Funding USD - Log Scale")
plt.ylabel("Number of Startups")
plt.show()
```

**Total funding over time**<br>
This graph shows how total startup funding changed over time since 1990. It helps identify years with stronger or weaker investment activity, and shows the sharp rise in overall funding in more recent years within the dataset.

```python
funding_rounds = dfs["funding_rounds"].copy()

funding_rounds["funded_at"] = pd.to_datetime(funding_rounds["funded_at"], errors="coerce")
funding_rounds["funded_year"] = funding_rounds["funded_at"].dt.year
funding_rounds["raised_amount_usd"] = pd.to_numeric(funding_rounds["raised_amount_usd"], errors="coerce")

funding_by_year = (
    funding_rounds
    .dropna(subset=["funded_year", "raised_amount_usd"])
    .query("funded_year >= 1990 and raised_amount_usd > 0")
    .groupby("funded_year")["raised_amount_usd"]
    .sum()
    .reset_index()
    .sort_values("funded_year")
)

plt.figure(figsize=(10,5))
sns.lineplot(data=funding_by_year, x="funded_year", y="raised_amount_usd", marker="o")
plt.title("Total Startup Funding Over Time Since 1990")
plt.xlabel("Year")
plt.ylabel("Total Funding USD")
plt.show()
```

**Number of Funding Rounds Over Time**<br>
This graph shows how many funding rounds occurred each year. It helps us see whether investment activity was increasing or slowing down over time.

```python
rounds_by_year = (
    funding_rounds
    .dropna(subset=["funded_year"])
    .query("funded_year >= 1990")
    .groupby("funded_year")
    .size()
    .reset_index(name="number_of_rounds")
    .sort_values("funded_year")
)

plt.figure(figsize=(10,5))
sns.lineplot(data=rounds_by_year, x="funded_year", y="number_of_rounds", marker="o")
plt.title("Number of Funding Rounds Over Time Since 1990")
plt.xlabel("Year")
plt.ylabel("Number of Funding Rounds")
plt.show()
```

### Cross-Cutting Analysis

**Status breakdown by category**<br>
This graph shows the percentage breakdown of company status (operating, acquired, closed, ipo) within each of the top 10 categories. This normalizes by category size so we can compare outcome rates fairly across industries with very different numbers of companies.

```python
status_by_category = (
    companies[companies["category_code"].isin(
        companies["category_code"].value_counts().head(10).index
    )]
    .groupby(["category_code", "status"])
    .size()
    .unstack(fill_value=0)
)

status_pct = status_by_category.div(status_by_category.sum(axis=1), axis=0) * 100

status_pct.plot(kind="bar", stacked=True, figsize=(10,6))
plt.title("Outcome Rate by Category (% of Companies, Top 10 Categories)")
plt.xlabel("Category")
plt.ylabel("% of Companies")
plt.xticks(rotation=45)
plt.legend(title="Status", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.show()
```

**Time-to-first-funding vs. outcome**<br>
This graph shows how many days passed between a company's founding and its first funding round, broken down by eventual status. It helps us see whether companies that got funded quickly had different outcomes than those that took longer.

```python
companies["days_to_first_funding"] = (
    companies["first_funding_at"] - companies["founded_at"]
).dt.days

valid = companies[
    (companies["days_to_first_funding"] >= 0) &
    (companies["days_to_first_funding"] < 3650)  # cap at 10 years to drop bad data
]

plt.figure(figsize=(10,5))
sns.boxplot(data=valid, x="status", y="days_to_first_funding")
plt.title("Days from Founding to First Funding, by Outcome")
plt.xlabel("Status")
plt.ylabel("Days to First Funding")
plt.show()
```

**Top investors**<br>
This graph shows the 15 most active investors by number of investments made. It helps us identify which firms appear most frequently across funding rounds in the dataset.

```python
investments = dfs["investments"].copy()
top_investors = (
    investments["investor_object_id"]
    .value_counts()
    .head(15)
)

# merge in names if investor_object_id maps to objects/funds
investor_names = dfs["objects"].set_index("id")["name"]
top_investors_named = top_investors.rename(index=investor_names)

plt.figure(figsize=(10,8))
top_investors_named.sort_values().plot(kind="barh")
plt.title("Top 15 Most Active Investors")
plt.xlabel("Number of Investments")
plt.ylabel("Investor")
plt.show()
```
