# EDA 
## Loading the data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

sns.set_theme(style="whitegrid")
pd.set_option("display.max_columns", 100)
```

```python
DATA_DIR = Path("/Users/leenisa/Documents/startup_ecosystem/data")
csv_files = list(DATA_DIR.glob("*.csv"))
csv_files
```

```python
dfs = {}

for file in csv_files:
    name = file.stem
    dfs[name] = pd.read_csv(file, low_memory=False)
    print(name, dfs[name].shape)
    
```

```python
for name, df in dfs.items():
    print("\n" + "-"*50)
    print(name)
    display(df.head())
```

## missing values 

```python
for name, df in dfs.items():
    print("\n" + "="*50)
    print(name)

    missing = (
        df.isna()
        .mean()
        .mul(100)
        .sort_values(ascending=False)
    )

    missing = missing[missing > 0]
    display(missing)
```

### exploring the main table: objects

```python
objects = dfs.get("objects")

objects.shape
```

```python
objects.head()
```

```python
objects.columns
```

```python
## creating a new table for the companies 

companies = objects.copy()

if "entity_type" in companies.columns:
    companies = companies[companies["entity_type"] == "Company"].copy()

companies.shape
```

```python
companies.head()
```

```python
date_cols = ["founded_at", "closed_at", "first_funding_at", "last_funding_at"]

for col in date_cols:
    if col in companies.columns:
        companies[col] = pd.to_datetime(companies[col], errors="coerce")

num_cols = ["funding_total_usd", "funding_rounds"]

for col in num_cols:
    if col in companies.columns:
        companies[col] = pd.to_numeric(companies[col], errors="coerce")

if "founded_at" in companies.columns:
    companies["founded_year"] = companies["founded_at"].dt.year
```

# GRAPHS

This graph shows the distribution of startup statuses.  
It helps us understand how many companies are operating, acquired, closed, or in other statuses.

```python
plt.figure(figsize=(8,5))
companies["status"].value_counts().plot(kind="bar")
plt.title("Startup Status Distribution")
plt.xlabel("Status")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.show()
```

# funded EDA

```python
funded = companies[
    (companies["funding_total_usd"] > 0) &
    (np.isfinite(companies["funding_total_usd"]))
].copy()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(funded["funding_total_usd"], bins=50)
plt.xscale("log")
plt.title("Distribution of Total Funding")
plt.xlabel("Total Funding USD - Log Scale")
plt.ylabel("Number of Startups")
plt.show()
```

```python
funding_rounds = dfs["funding_rounds"].copy()

funding_rounds["funded_at"] = pd.to_datetime(funding_rounds["funded_at"], errors="coerce")
funding_rounds["funded_year"] = funding_rounds["funded_at"].dt.year
funding_rounds["raised_amount_usd"] = pd.to_numeric(funding_rounds["raised_amount_usd"], errors="coerce")

funding_by_year = (
    funding_rounds
    .dropna(subset=["funded_year", "raised_amount_usd"])
    .query("funded_year >= 1990 and raised_amount_usd > 0")
    .groupby("funded_year")["raised_amount_usd"]
    .sum()
    .reset_index()
    .sort_values("funded_year")
)

plt.figure(figsize=(10,5))
sns.lineplot(data=funding_by_year, x="funded_year", y="raised_amount_usd", marker="o")
plt.title("Total Startup Funding Over Time Since 1990")
plt.xlabel("Year")
plt.ylabel("Total Funding USD")
plt.show()
```

```python
rounds_by_year = (
    funding_rounds
    .dropna(subset=["funded_year"])
    .query("funded_year >= 1990")
    .groupby("funded_year")
    .size()
    .reset_index(name="number_of_rounds")
    .sort_values("funded_year")
)

plt.figure(figsize=(10,5))
sns.lineplot(data=rounds_by_year, x="funded_year", y="number_of_rounds", marker="o")
plt.title("Number of Funding Rounds Over Time Since 1990")
plt.xlabel("Year")
plt.ylabel("Number of Funding Rounds")
plt.show()
```

```python
funding_rounds["funded_at"] = pd.to_datetime(funding_rounds["funded_at"], errors="coerce")
funding_rounds["funded_year"] = funding_rounds["funded_at"].dt.year
funding_rounds["raised_amount_usd"] = pd.to_numeric(funding_rounds["raised_amount_usd"], errors="coerce")
```

This graph shows how total startup funding changed over time.  
It can help identify years with stronger or weaker investment activity.

```python
funding_by_year = (
    funding_rounds
    .dropna(subset=["funded_year"])
    .groupby("funded_year")["raised_amount_usd"]
    .sum()
    .reset_index()
)

plt.figure(figsize=(10,5))
sns.lineplot(data=funding_by_year, x="funded_year", y="raised_amount_usd", marker="o")
plt.title("Total Startup Funding Over Time")
plt.xlabel("Year")
plt.ylabel("Total Funding USD")
plt.show()
```

```python
plt.figure(figsize=(10,5))
funding_rounds["funding_round_type"].value_counts().head(15).plot(kind="bar")
plt.title("Most Common Funding Round Types")
plt.xlabel("Funding Round Type")
plt.ylabel("Number of Rounds")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_funding_type = (
    funding_rounds
    .dropna(subset=["funding_round_type", "raised_amount_usd"])
    .groupby("funding_round_type")["raised_amount_usd"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
avg_funding_type.plot(kind="bar")
plt.title("Average Funding Amount by Round Type")
plt.xlabel("Funding Round Type")
plt.ylabel("Average Raised Amount USD")
plt.xticks(rotation=45)
plt.show()
```

This graph compares funding between different startup statuses.  
The log scale is used because funding values are very spread out.

```python
plt.figure(figsize=(9,5))
sns.boxplot(data=funded, x="status", y="funding_total_usd")
plt.yscale("log")
plt.title("Funding Distribution by Startup Status")
plt.xlabel("Status")
plt.ylabel("Total Funding USD - Log Scale")
plt.xticks(rotation=45)
plt.show()
```

# Acquisitions EDA

```python
acquisitions = dfs.get("acquisitions")

acquisitions.head()
```

```python
acquisitions["acquired_at"] = pd.to_datetime(acquisitions["acquired_at"], errors="coerce")
acquisitions["acquired_year"] = acquisitions["acquired_at"].dt.year
acquisitions["price_amount"] = pd.to_numeric(acquisitions["price_amount"], errors="coerce")
```

```python
acq_by_year = acquisitions.groupby("acquired_year").size().reset_index(name="count")

plt.figure(figsize=(10,5))
sns.lineplot(data=acq_by_year, x="acquired_year", y="count", marker="o")
plt.title("Number of Acquisitions Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Acquisitions")
plt.show()
```

# IPO EDA

```python
ipos = dfs.get("ipos")

ipos.head()
```

```python
ipos["public_at"] = pd.to_datetime(ipos["public_at"], errors="coerce")
ipos["ipo_year"] = ipos["public_at"].dt.year
```

```python
ipo_by_year = ipos.groupby("ipo_year").size().reset_index(name="count")

plt.figure(figsize=(10,5))
sns.lineplot(data=ipo_by_year, x="ipo_year", y="count", marker="o")
plt.title("Number of IPOs Over Time")
plt.xlabel("Year")
plt.ylabel("Number of IPOs")
plt.show()
```
